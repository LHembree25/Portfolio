After running into paywall issues with OpenAI API, I used BERT (Bidirectional Encoder Representations from Transformers) which is a Natural Language Processing (NLP) model developed by Google. It can be used for translation, text generation, and text summarization. Medium.com classifies it as a LLM so it fits into the parameters of the assignment, most at least since we are supposed to use OpenAI API. The objective was to pull text from the invoice and convert it into structured JSON. 

```python
# Import library
import pdfplumber

# Read the invoice
invoice_path = "C:/Users/Ashem/OneDrive/Desktop/Masters Program - DSC/DSC 670 - Advanced Uses of Generative AI/Week 2/dsc-670-exercise-invoice.pdf"

# Extract the text
with pdfplumber.open(invoice_path) as pdf:
    extracted_text = "\n".join(page.extract_text() for page in pdf.pages)

# ensure it read
print(extracted_text)
```

I used pdfplumber for extraction because it gives direct access to page text. The extraction step produced a clean text block with all invoice fields visible. I used that model to identify entities in the extracted text. The model tagged locations, organizations, and person names. I need the model to be able to read the raw text it extracts so I need to run it through BERT so it gets a structured format it can do that with. (I'd do way more Bert and Ernie jokes but I can't remember most of sesame street at this point)

```python
# Import library
from transformers import AutoTokenizer, AutoModelForTokenClassification
import torch

# Load BERT from local area
tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")
```

```python
# Run BERT (Ernie is missing)
inputs = tokenizer(extracted_text, return_tensors="pt", truncation=True)
outputs = model(**inputs)
preds = torch.argmax(outputs.logits, dim=2)
tokens = inputs.tokens()
labels = [model.config.id2label[p.item()] for p in preds[0]]
```

After never using BERT before and getting through with minimal network issues (own system errors) that prevented the code from running, the system ran without error and allowed the JSON code to write an output. 

```python
# Import library
import json

# Convert to JSON
entities = []
for token, label in zip(tokens, labels):
    if label != "O":
        entities.append({"token": token, "label": label})

json_output = json.dumps({"entities": entities}, indent=4)
print(json_output);
```

The BERT model performed as expected for a general Named Entity Recognition (NER) model, it identified names and locations correctly. What it did not identify, were invoice‑specific fields such as totals, dates, or line items because the model is not trained for financial document structure. This is a limitation of the model rather than the extraction process; this is the shortfall I would need to determine a path forward in dealing with during other applications of this output. Otherwise the extracted text was complete, with the bottleneck unfortunately being the model’s domain knowledge.

The JSON output reflects the entities the model was able to detect. The structure is valid and consistent. The output is not a full invoice schema, but it is a structured representation of the model’s interpretation of the text. This aligns with the assignment requirement to convert extracted text into structured JSON using an LLM.
