Ch 7 transformations

- Filter out missing data
- bin by Accession Year (not actually, went with manipulating the strings instead for reasons later explained)

```python
# Check out the data
metobj_df.info
```

```python
# Determine how many na values exist
metobj_df.isna().sum()
```

```python
# drop the na values
metobj_df.dropna()
```

```python
# determine if we can bin by Accession Year
metobj_df["AccessionYear"].unique()
```

```python
# Values aren't clean so we have to manipulate the string to clean them
# strip whitespace
metobj_df.columns = metobj_df.columns.str.strip()
```

```python
# Convert to string
col = metobj_df["AccessionYear"].astype(str).str.strip()

# extract the year from each value
years = col.str.extract(r"(\d{4})")[0]

# ensure the years are numeric
metobj_df["AccessionYear_clean"] = pd.to_numeric(years, errors="coerce")

# injection due to noticing a stray "." after my values & it bugged me
metobj_df["AccessionYear_clean"] = (
    metobj_df["AccessionYear_clean"].astype("Int64")
)

# no need to bin if I did the manipulate strings part of Ch7
```

```python
# relook at Accession Year
metobj_df["AccessionYear_clean"].unique()
```

```python
# apply the grouping
metobj_df["Dept_Group"] = metobj_df["Department"].map(Dept_Groups)
```
